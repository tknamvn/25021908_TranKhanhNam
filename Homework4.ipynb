```python
#W4A9
n = int(input("Nhap n: "))
cp = ""
def kt(k):
    ktra = True
    k = str(k)
    K = list(k)
    for i in range(len(K)-1):
        if K[i] == K[i + 1]:
            ktra = False
            break
    if ktra == True:
        return True
    else:
        return False
count = 0
for s in range(4, n +1):
    if kt(s) == True and int(s**(0.5))**2 == int(s):
        cp += str(s) + " "
        count += 1
if count == 0:
    print("No number")
else:
    print(str(cp))
```

```python
#W4A10
n = int(input("Nhap n: "))
xln = 0
def collatz(k):
    count = 1
    while k > 1:
        if k % 2 == 0:
            k = int(k/2)
            count += 1
        else:
            k = 3*k + 1
            count += 1
    return count
x = 0
so = 0
ans = ""
dayc = {}
for i in range(1, n+1):
    x = collatz(i)
    if x > xln:
        xln = x
    if x not in dayc:
        dayc[x] = []
    dayc[x].append(i)
so = min(dayc[xln])
ans = str(so) + " " + str(xln)
print(ans)
```

```python
#W4A13
a, b = map(int, input("Nhap 2 so a va b: ").split())
def tuoc(n):
    tong = 1
    for k in range(2, int(n**(0.5)+1)):
        if n % k == 0:
            tong += k + n//k
    return tong
if tuoc(a) == b and tuoc(b) == a:
    print("true")
else:
    print("false")
```
