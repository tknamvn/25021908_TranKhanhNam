```python
#W5A3
import math
n=int(input())
def ktsnt(n):
    if n==1:
        return False
    else:
        for i in range(2,math.isqrt(n)+1):
            if n%i==0:
                return False
    return True
print(ktsnt(n))
```

```python
#W5A4
n=int(input())
def ktshh(x):
    tong=0
    for i in range(1,n//2+1):
        if x%i==0:
            tong+=i
    if tong==x:
        return True
    else:
        return False
print(ktshh(n))
```

```python
#W5A9
n=int(input())

def tcs(x):
    tong=0
    while x>0:
        tong+= x%10
        x=x//10
    return tong
    
print(tcs(n))
```

```python
#W5A10
a,b=map(str,input().split())
def ktax(a,b):
    m = {} 
    s = set() 
    for char_a, char_b in zip(a, b):
        if char_a in m:
            if m[char_a] != char_b:
                return False
        else:
            if char_b in s:
                return False
            m[char_a] = char_b
            s.add(char_b)
    return True
print(ktax(a,b))
```
