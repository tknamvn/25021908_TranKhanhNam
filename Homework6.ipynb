```python
#W6A1
N=list(map(int,input().split()))
def remove_dup(N):
    seen = set()
    result = []
    
    for x in N:
        if x not in seen:
            result.append(x)
            seen.add(x)
    
    return result
print(remove_dup(N))
            
            
```

```python
#W6A3
a=tuple(eval(input()))
k=int(input())
def xoay(N, k):
    n = len(N)
    if n == 0 or k == 0:
        return N
        
    dich = k % n
    rotated = N[dich:] + N[:dich]
    return rotated

kq = xoay(a,k)
print(kq)
```

```python
#W6A9
import ast
from collections import defaultdict

def get_dict_input():
    input_str = input()
    return ast.literal_eval(input_str)

dicA = get_dict_input()
dicB = get_dict_input()
dicC = defaultdict(int)

for key, value in dicA.items():
    if isinstance(value, (int, float)):
        dicC[key] += value

for key, value in dicB.items():
    if isinstance(value, (int, float)):
        dicC[key] += value

sorted_result = dict(sorted(dicC.items()))
print(sorted_result)
```

```python
#W6A10
N=list(map(int,input().split()))
k=int(input())
A=[]
for i in range(len(N)-1):
    for j in range(i+1,len(N)):
        if N[i]+N[j]==k:
            A.append((N[i],N[j]))
            break
print(A)
```
