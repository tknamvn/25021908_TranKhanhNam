```python
#W6A1
N=list(map(int,input().split()))
def remove_dup(N):
    seen = set()
    result = []
    
    for x in N:
        if x not in seen:
            result.append(x)
            seen.add(x)
    
    return result
print(remove_dup(N))
            
            
```

```python
#W6A3
a=tuple(eval(input()))
k=int(input())
def xoay(N, k):
    n = len(N)
    if n == 0 or k == 0:
        return N
        
    dich = k % n
    rotated = N[dich:] + N[:dich]
    return rotated

kq = xoay(a,k)
print(kq)
```

```python
#W6A4
N=list(map(str,input().split()))
dic={}
for i in N:
    k,v=i.split(":")
    if k in dic:
        dic[k].append(v)
    else:
        dic[k]=[v]
print(dic)
    
```

```python
#W6A5
N=list(map(int,input().split()))
dic={
    "positives":0,
    "negatives":0,
    "zeros":0
}
for i in N:
    if i>0:
        dic["positives"]+=1
    elif i<0:
        dic["negatives"]+=1
    else:
        dic["zeros"]+=1
print(dic)
```

```python
#W6A8
N=list(map(str,input().split()))
m=set(N)
dic={}
for i in m:
    dem=0
    for k in range(len(N)):
        if N[k]==i:
            dem+=1
    dic[i]=dem
print(dic)
```

```python
#W6A9
import ast
from collections import defaultdict

def get_dict_input():
    input_str = input()
    return ast.literal_eval(input_str)

dicA = get_dict_input()
dicB = get_dict_input()
dicC = defaultdict(int)

for key, value in dicA.items():
    if isinstance(value, (int, float)):
        dicC[key] += value

for key, value in dicB.items():
    if isinstance(value, (int, float)):
        dicC[key] += value

sorted_result = dict(sorted(dicC.items()))
print(sorted_result)
```
