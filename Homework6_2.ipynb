```python
#W6A28
N=list(map(int,input().split()))

def kttang(N):
    for i in range(len(N)-1):
        if N[i]>N[i+1]:
            return False
    return True
    
def ktgiam(N):
    for i in range(len(N)-1):
        if N[i]<N[i+1]:
            return False
    return True
    
if kttang(N)==True:
    print("Day tang thuc su")
elif ktgiam(N)==True:
    print("Day giam thuc su")
else:
    print("Day vo danh")
```
